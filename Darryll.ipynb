FOLD 3

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
df3 = pd.read_csv(r'C:\GDSCHackathon\im-somewhat-of-a-cybersecurity-analyst-myself\training\uploads\fold_3.csv')
```

```python
nan_cols30 = [i for i in df3.columns if df3[i].isnull().sum() > 0.30*len(df3)]
```

```python
df3.drop(nan_cols30, axis = 1, inplace = True)
```

```python
bool_cols = [col for col in df3 
             if np.isin(df3[col].dropna().unique(), [0, 1]).all()]
```

```python
for i in bool_cols:
    df3[i].fillna(df3[i].mode()[0], inplace=True)
```

```python
df3 = df3.fillna(0)
```

```python
for col in df3:
    all_names = set(df3[col])
    mapper = {k: 0 + i for i, k in enumerate(all_names )}
    df3[col] = df3[col].map(mapper)

df3
```
